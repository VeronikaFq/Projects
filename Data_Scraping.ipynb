```python
import requests
from bs4 import BeautifulSoup
import urllib
```

```python
def get_actors_by_movie_soup(cast_page_soup, num_of_actors_limit=None):
    url = ''
    actors = cast_page_soup.find_all('tr', attrs={'class':['odd', 'even']})
    all_actors_list = []
    for actor in actors:
        name_of_actor = list(list(list(actor.children)[3].children)[1])[0][1:-1]
        link = list(list(actor.children)[3].children)[1]['href']
        url_to_actor_page = urllib.parse.urljoin('https://www.imdb.com', link)
        all_actors_list.append((name_of_actor, url_to_actor_page))
    if num_of_actors_limit==None:
        return all_actors_list
    else:
        if len(all_actors_list)<num_of_actors_limit:
            return all_actors_list
        else: 
            return all_actors_list[:num_of_actors_limit]
        
def get_movies_by_actor_soup(actor_page_soup, num_of_movies_limit=None):
    films = actor_page_soup.find_all('div', attrs={'class':'filmo-category-section'})
    films_list = []
    for film in list(films[0].children):
        if film.text != '\n' and len(list(film.children)[4].text) <= 1  and len(list(film.children)[5].text) == 0:
            name_of_movie = list(film.children)[3].text
            link = list(list(film.children)[3].children)[0]['href']
            url_to_movie_page = urllib.parse.urljoin('https://www.imdb.com', link)
            films_list.append ((name_of_movie, url_to_movie_page))    
    
    if num_of_movies_limit==None:
        return films_list
    else:
        if len(films_list)<num_of_movies_limit:
            return films_list
        else: 
            return films_list[:num_of_movies_limit]
```

```python
#Output 1: movie "007:No Time to Die"

url_1 = 'https://www.imdb.com/title/tt2382320/fullcredits/'
headers = {'Accept-Language':'en',
          'X-FORWARDED_FOR': '2.21.184.0'}
response_1 = requests.get(url_1, headers = headers)
soup_1 = BeautifulSoup(response_1.text)

get_actors_by_movie_soup(soup_1)
```

```python
#Output 2: movie "Black Widow"

url_2 = 'https://www.imdb.com/title/tt3480822/fullcredits/'
headers = {'Accept-Language':'en',
          'X-FORWARDED_FOR': '2.21.184.0'}
response_2 = requests.get(url_2, headers = headers)
soup_2 = BeautifulSoup(response_2.text)

get_actors_by_movie_soup(soup_2)
```

```python
# Output 3: movie "Black Widow", num_of_movies_limit=150

get_actors_by_movie_soup(soup_2, 150)
```

```python
# Output 4: movie "Black Widow", num_of_movies_limit=5

get_actors_by_movie_soup(soup_2, 5)
```
